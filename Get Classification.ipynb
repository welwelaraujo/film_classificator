```python
dfList = pd.read_csv(r"C:\\Users\\drt44888\\Documents\\Git\\film_classificator\\data\\Tabela_Obras_Cinema.txt", sep = ";", encoding='latin-1')
vecRange = random_list(len(dfList),130)
dfSelected = dfList.iloc[vecRange]
```

```python
browser = webdriver.Chrome()
titulo = []
faixa = []
```

```python
for element in range(0,len(dfSelected)):
    try:
        print("Bucando a classificação " + str(element))
        dfSelected["Titulo no Brasil"].iloc[element] 
        browser.get("http://portal.mj.gov.br/ClassificacaoIndicativa/jsps/ConsultarObraForm.do?combo_action")
        browser.find_element_by_xpath("/html/body/table/tbody/tr/td/table[6]/tbody/tr/td/form[1]/table[1]/tbody/tr[1]/td[2]/input").send_keys(dfSelected["Titulo no Brasil"].iloc[element])
        browser.find_element_by_xpath("/html/body/table/tbody/tr/td/table[6]/tbody/tr/td/form[1]/table[1]/tbody/tr[3]/td[2]/input").send_keys(dfSelected["Titulo Original"].iloc[element])
        browser.find_element_by_xpath("/html/body/table/tbody/tr/td/table[6]/tbody/tr/td/form[1]/table[2]/tbody/tr/td/a/img").click()
        browser.find_element_by_xpath('//*[@id="lista"]/tbody/tr/td[5]/a/img').click()
        title = browser.find_element_by_xpath('//*[@id="TRbl_report_ClassificacaoProcessoObraView1"]/td[6]/font').text
        browser.find_element_by_xpath("/html/body/table/tbody/tr/td/table[6]/tbody/tr/td/a/img").click()
        titulo.append(dfSelected["Titulo no Brasil"].iloc[element])
        faixa.append(title)
    except:
        pass
    browser.get("http://portal.mj.gov.br/ClassificacaoIndicativa/jsps/ConsultarObraForm.do?combo_action")
    browser.delete_all_cookies()
    browser.refresh()
```

```python
dfFilms = {"titulo":titulo, "faixa_etaria":faixa}
dfFilms = pd.DataFrame(dfFilms)
```

```python
def random_list(n_max,n):
    randomVect = []
    while(len(randomVect) <= n-1):
        number = randint(0,n_max)
        try:
            randomVect.index(number)
            pass
        except:
            randomVect.append(number)
    return randomVect
```
