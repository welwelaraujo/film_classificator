```python
from selenium import webdriver
import pandas as pd
```

```python
corpus = pd.read_csv(r"data\sinopses.csv")
corpus['titulo'] = [x.strip().replace(':', ' -') for x in corpus['titulo']]
```

```python
browser = webdriver.Chrome()
titulo = []
faixa = []
```

```python
for element in range(0,len(corpus)):
    try:
        print("Bucando a classificação " + str(element))
        corpus["titulo"].iloc[element] 
        browser.get("http://portal.mj.gov.br/ClassificacaoIndicativa/jsps/ConsultarObraForm.do?combo_action")
        browser.find_element_by_xpath("/html/body/table/tbody/tr/td/table[6]/tbody/tr/td/form[1]/table[1]/tbody/tr[1]/td[2]/input").send_keys(corpus["titulo"].iloc[element])
        browser.find_element_by_xpath('//*[@id="comboCategoria"]/option[32]').click()
        browser.find_element_by_xpath("/html/body/table/tbody/tr/td/table[6]/tbody/tr/td/form[1]/table[2]/tbody/tr/td/a/img").click()
        browser.find_element_by_xpath('//*[@id="lista"]/tbody/tr/td[5]/a/img').click()
        title = browser.find_element_by_xpath('//*[@id="TRbl_report_ClassificacaoProcessoObraView1"]/td[6]/font').text
        browser.find_element_by_xpath("/html/body/table/tbody/tr/td/table[6]/tbody/tr/td/a/img").click()
        titulo.append(corpus["titulo"].iloc[element])
        faixa.append(title)
    except:
        titulo.append(corpus["titulo"].iloc[element])
        faixa.append("vazio")
        pass
    browser.get("http://portal.mj.gov.br/ClassificacaoIndicativa/jsps/ConsultarObraForm.do?combo_action")
    browser.delete_all_cookies()
    browser.refresh()
browser.close()
```

```python
dfFilms = {"titulo":titulo, "faixa_etaria":faixa}
dfFilms = pd.DataFrame(dfFilms)
```

```python
dfFilms['sinopse'] =  corpus['sinopse']
```

```python
for i in range(0,len(dfFilms)):
    try:
        n_cut = str(dfFilms["faixa_etaria"].iloc[i]).index("1")
        dfFilms["faixa_etaria"][i] = str(dfFilms["faixa_etaria"].iloc[i])[n_cut:n_cut+2]
    except ValueError:
        try:
            n_cut = str(dfFilms["faixa_etaria"].iloc[i]).index("livre")
            dfFilms["faixa_etaria"][i] = "livre"
        except ValueError:
            pass
```

```python
dfFilms.to_csv("data\corpus.csv",index=False)
```
